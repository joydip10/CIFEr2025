```python
def prepare_dataset(ticker, year):
    
    #download dataset
    start_date = datetime.datetime(year, 1, 1)
    end_date = datetime.datetime(year, 12, 31)
    data = yf.download(ticker, start=start_date, end=end_date)
    
    
    #standardization
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)
    
    #dimensionality reduction using KPCA
    kpca = KernelPCA(n_components=1, kernel='cosine', gamma=15)
    data_kpca_reduced = kpca.fit_transform(data_scaled)
    data_kpca_reduced=pd.DataFrame(data_kpca_reduced)
    
    #rename the reduced feature
    data_kpca_reduced=data_kpca_reduced.rename(columns={0:f'{ticker}'})
    data_kpca_reduced.index=data.index
    
    plt.subplot(2,1,1)
    #plot Adjusted Closing price
    plt.title(f'{ticker} True Adjusted Closing Price')
    plt.plot(data.index,data['Adj Close'],label=f'{ticker} Adj Close')
    plt.legend()
    
    plt.show()
    
    plt.subplot(2,1,2)
    #plot dimensionality reduced dataset
    plt.title(f'{ticker} Reduced Feature')
    plt.plot(data_kpca_reduced.index,data_kpca_reduced[f'{ticker}'],label=f'reduced {ticker} feature')
    plt.legend()
    plt.show()
    
    return data_kpca_reduced
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'JPM', 'BAC', 'V', 'PG', 'KO', 'WMT', 'PEP', 'CVX', 'XOM', 'SLB', 'BA', 'UPS', 'CAT', 'MMM', 'JNJ', 'PFE', 'MRNA', 'DIS', 'NFLX', 'T', 'NVDA', 'ORCL', 'INTC', 'LMT', 'GS', 'COST', 'BTC-USD', 'ETH-USD', 'LTC-USD', 'XRP-USD']
```

```python
dataset_2020=[]
dataset_2021=[]
dataset_2022=[]
dataset_2023=[]
```

```python
for i in tickers:
    dataset_2020.append(prepare_dataset(i,2020))
    dataset_2021.append(prepare_dataset(i,2021))
    dataset_2022.append(prepare_dataset(i,2022))
    dataset_2023.append(prepare_dataset(i,2023))
```

```python
def merge_stocks_and_cryptos(dataset):
    
    # getting stocks and cryptos separately
    stocks = dataset[:-4]  
    cryptos = dataset[-4:] 
    
    # Dropping the weekends from cryptos
    refined_cryptos = []
    for crypto in cryptos:
        crypto_no_weekends = crypto[crypto.index.weekday < 5]  # Keep only weekdays
        refined_cryptos.append(crypto_no_weekends)

    # merging stocks and cryptos together
    merged_stocks = pd.concat(stocks, axis=1)
    merged_cryptos = pd.concat(refined_cryptos, axis=1)
    merged_data = pd.concat([merged_stocks, merged_cryptos], axis=1)

    #dropping the remaining null values (especially due to cryptos- holiday data)
    merged_data.dropna(inplace=True)
    
    return merged_data
```

```python
data_2020=merge_stocks_and_cryptos(dataset_2020)
data_2021=merge_stocks_and_cryptos(dataset_2021)
data_2022=merge_stocks_and_cryptos(dataset_2022)
data_2023=merge_stocks_and_cryptos(dataset_2023)
```

```python
data_2020.pct_change().dropna().to_csv('data_2020.csv',index=False)
data_2021.pct_change().dropna().to_csv('data_2021.csv',index=False)
data_2022.pct_change().dropna().to_csv('data_2022.csv',index=False)
data_2023.pct_change().dropna().to_csv('data_2023.csv',index=False)
```
